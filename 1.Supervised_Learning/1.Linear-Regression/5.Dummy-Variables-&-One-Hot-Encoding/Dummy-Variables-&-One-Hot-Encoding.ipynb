```python
import pandas as pd
df = pd.read_csv("./homeprices.csv")
df
```

```python
# get dummies for the required column using pandas
dummies = pd.get_dummies(df.town).astype(int) # .astype(int) used to display the values in integer form (i.e., 0 and 1)
dummies
```

```python
merged = pd.concat([df,dummies], axis = 'columns')
merged
```

```python
final = merged.drop(["town", "west windsor"], axis = 'columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
x = final.drop("price", axis = "columns") # removing dependent variable
x
```

```python
y = final.price
y
```

```python
model.fit(x,y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

```python
model.score(x,y)
```
